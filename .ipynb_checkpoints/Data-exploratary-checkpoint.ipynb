## (0) Import Library and Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.graph_objects as go
sb.set()
```

```python
customerData = pd.read_csv('dataset\marketing_campaign.csv',sep='\t')
```

We also set up some of the utility function, which will help with the visualization later.

```python
def histogramCountOld(dataset,x,axes = [],i = 0,j = 0,meta = {"height": 4.8, "width": 6.4}):
    
    import seaborn as sb
    
    f = plt.figure(figsize=(meta["height"],meta["width"]))
    
    #Cite M2 ExploratoryAnalysis
    print("Number of", x ,":", len(dataset[x].unique()))
    
    print(dataset[x].value_counts())
    if(len(axes) != 0):
        sb.countplot(y=x, data=dataset, ax = axes[i,j], orient = "h")
    else:
        sb.catplot(y = x, data = dataset, kind = "count")
```

```python
def histogramCountSubplot(dataset,xArray,x_count,y_count,meta = {"height" : 20, "width":12}):
    
    import seaborn as sb
    import matplotlib.pyplot as plt
    
    if(x_count * y_count != len(xArray)):
        raise IndexError("Graph count does not match with the number of response.")
        
    f, axes = plt.subplots(y_count, x_count, figsize=(meta["width"], meta["height"]))

    i = 0
    j = 0
    for var in xArray:
        histogramCountOld(customerData,var,axes,i,j)
        print("")
        j+=1
        if(j == x_count):
            j=0
            i+=1
```

```python
def plotlyDist(dataset,x,binSize,meta = {"height" : 500}):
    
    import plotly.graph_objects as go
    import plotly.figure_factory as ff
    from plotly.subplots import make_subplots
    
    fig0 = ff.create_distplot([dataset[dataset[x].notna() == True][x].values], [x],bin_size = binSize,curve_type = "kde",histnorm = "")

    fig = make_subplots(rows=2, cols=1,
                        subplot_titles=(x,""),
                       row_heights=[0.2,0.8],
                       vertical_spacing=0.02)

    box0 = go.Box(x = dataset[x],legendgroup = x,showlegend = False,name = "")
    trace0 = go.Histogram(fig0['data'][0],legendgroup = x,marker_color='DarkOrchid')
    line0 = go.Scatter(fig0['data'][1],legendgroup = x,line=dict(color='RebeccaPurple',width=1))

    fig.append_trace(box0, 1, 1)
    fig.append_trace(trace0, 2, 1)
    fig.append_trace(line0, 2, 1)


    fig.update_layout(height=meta["height"])

    fig.show()
```

```python
def numVarDetails(dataset,x,meta = {"height" : 5,"width" : 15}):
    
    import seaborn as sb
    # Cite M2 BasicStatistics
    # Set up matplotlib figure with three subplots

    f, axes = plt.subplots(1, 3, figsize=(meta["width"], meta["height"]))

    # Plot the basic uni-variate figures for HP
    sb.boxplot(data = dataset[x], orient = "h", ax = axes[0])
    sb.histplot(data = dataset[x], ax = axes[1])
    sb.violinplot(data = dataset[x], orient = "h", ax = axes[2])
```

Based on the description on the dataset, we can convert some of the variables' data types to categorical, to help with the exploratary. 

```python
colOfNum = []
colOfCat = []
for col in customerData:
    if str(customerData[col].dtypes) != "int64" and str(customerData[col].dtypes) != "float64":
        colOfCat.append(str(col))
    else:
        colOfNum.append(str(col))
        
colOfCat.extend(['Complain','Kidhome','Teenhome','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response'])

for element in ['Complain','Kidhome','Teenhome','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']:
    colOfNum.remove(element)
    
print("Column which consider as Numerical Data:\n", colOfNum)
print()
print("Column which consider as Categorical Data:\n",colOfCat)
```

```python
for data in colOfCat:
    customerData[data] = customerData[data].astype("category")
```

```python
customerData[colOfNum].describe()
```

```python
customerData[colOfCat].describe()
```

### <a name="Education"></a>2   Education

There are 5 categories in Education, the categories themselves are self-explanatory other than 2n Cycle. 

After some research, we find out that 2n cycle a qualification given in European countries, from the sources, they are equivalent to Master degree. [Link](http://www.ehea.info/page-three-cycle-system)  

At the end, we decide not to combine Master and 2n Cycle as they may still give different meaning, for example, this can mean people obtain their qualification from different area (e.g. getting 2nd Cycle meaning they get their degree in Europe, getting Master meaning they get their degree in places other than Europe).

```python
histogramCountOld(customerData,"Education")
```
